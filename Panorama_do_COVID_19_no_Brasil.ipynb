# **Panorama do COVID-19 no Brasil em Junho de 2020**

COVID-19 (do inglês Coronavirus Disease 2019) é uma doença infecciosa causada pelo novo coronavírus da síndrome respiratória aguda grave 2 (SARS-CoV-2) identificado pela primeira vez em seres humanos em Wuhan, na China, em Dezembro de 2019. 

Os sintomas mais comuns são febre, tosse seca e cansaço, porém outros sintomas menos comuns são dores musculares, dor de garganta, dor de cabeça, congestão nasal, conjuntivite, perda do olfato e do paladar, erupções cutâneas na pele ou dedos de tom azul.

A maior dos casos, cerca de 80%, são assintomáticos e a maioria recupera sem sequelas. Entretanto, 15% são infeções graves que necessitam de oxigênio e 5% são infecções muito graves que necessitam de ventilação assistida em unidade de terapia intensiva (UTI) de hospital.

A doença transmite-se através de gotículas produzidas nas vias respiratórias das pessoas infectadas. Ao tossir ou espirrar, estas gotículas podem ser inaladas ou atingir diretamente o nariz, a boca ou olhos de pessoas próximas. 

O intervalo de tempo entre a exposição ao vírus e o início dos sintomas é de 2 a 14 dias, sendo em média de 5 dias.

Entre as medidas de prevenção estão a lavagem frequente das mãos, distanciamento de mais de 1,5 m e máscaras. Ainda não existe vacina aprovada mas os testes em humanos já tiveram início. Tratamentos estão sendos testados mas nenhum com comprovada eficácia. Consequentemente, muitos governos optaram por implantar diversos tipos de quarentenas para aumentar o distanciamento social. O efeito colateral claro é a redução drástica de consumo resultando em queda de atividade econômica e altas taxas de desemprego.

O objetivo deste artigo é analisar os dados públicos da COVID-19 no Brasil e compatilhar as conclusões.

Nações Unidas, Banco Mundial, Carga Global de Doenças, Escola de Governo Blavatnik, etc. A dificuldade de se obter dados em vários locais.

No Brasil utilizamos os dados do site [Coronavirus Brasil](https://https://covid.saude.gov.br/). Recentemente o governo federal começou a dificultar a divulgação de dados da COVID-19. Isso levou a criação de um  consórcio de veículos de imprensa. Outra notícia divulgada foi um desencontro de informações entre os dados do governo paulista e da prefeitura de São Paulo. ### **Dicionário de Variáveis**
* **iso_code**: Códigos de países com três letras
* **continent**: Continente da localização geográfica
* **location**: Localização geográfica
* **date**: Data da observação
* **total_cases**: Total de casos confirmados de COVID-19
* **new_cases**: 
Novos casos confirmados de COVID-19
* **total_deaths**: 
Total de mortes atribuídas ao COVID-19
* **new_deaths**: 
Novas mortes atribuídas ao COVID-19
* **total_cases_per_million**: 
Total de casos confirmados de COVID-19 por 1.000.000 de pessoas
* **new_cases_per_million**: 
Novos casos confirmados de COVID-19 por 1.000.000 de pessoas
* **total_deaths_per_million**: 
Total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas
* **new_deaths_per_million**: 
Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas
* **total_tests**: 
Total de testes para COVID-19
* **new_tests**: 
Novos testes para COVID-19
* **new_tests_smoothed**: Novos testes para COVID-19 (suavizado por 7 dias). Para países que não relatam dados de teste diariamente, assumimos que o teste mudou igualmente diariamente em qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que são calculados em média em uma janela de 7 dias consecutivos
* **total_tests_per_thousand**: 
Total de testes para COVID-19 por 1.000 pessoas
* **new_tests_per_thousand**: 
Novos testes para COVID-19 por 1.000 pessoas
* **new_tests_smoothed_per_thousand**: 
Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas
* **tests_units**: 
Unidades usadas pelo local para relatar seus dados de teste
* **stringency_index**: Índice de Estresse da Resposta do Governo: medida composta baseada em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens, redimensionados para um valor de 0 a 100 (100 = resposta mais estrita)
* **population**: 
População em 2020
* **population_density**: 
Número de pessoas divididas por área terrestre, medida em quilômetros quadrados, no ano mais recente disponível
* **median_age**: 
Idade média da população, projeção da ONU para 2020
* **aged_65_older**: 
Percentual da população com 65 anos ou mais, ano mais recente disponível
* **aged_70_older**: 
Percentual da população com 70 anos ou mais em 2015
* **gdp_per_capita**: 
Produto interno bruto com paridade do poder de compra (dólares internacionais constantes de 2011), último ano disponível
* **extreme_poverty**: 
Percentual da população que vive em extrema pobreza, último ano disponível desde 2010
* **cvd_death_rate**: 
Taxa de mortalidade por doenças cardiovasculares em 2017
* **diabetes_prevalence**: 
Prevalência de diabetes (% da população entre 20 e 79 anos) em 2017
* **female_smokers**: 
Percentual de mulheres que fumam, último ano disponível
* **male_smokers**: 
Percentual de homens que fumam, último ano disponível
* **handwashing_facilities**: 
Proporção da população com instalações básicas de lavagem das mãos nas instalações, último ano disponível
* **hospital_beds_per_thousand**: 
Camas de hospital por 1.000 pessoas, último ano disponível desde 2010
* **life_expectancy**: Expectativa de vida no nascimento em 2019

* Por conta disso, fizemos uma matriz de correlação sem considerar dados que  tenham mais de 50% de valores nulos.

* Abaixo podemos ver um mapa de correlação. Considerando as variáveis que têm mais correlação com morte por milhão, identificamos que países com alta média de idade e principalmente com mais de 65 anos tem maior correlação com alta taxas de morte por milhão. 

* Obviamente um maior número de casos leva a uma maior número de mortes. Entretanto, um dado curioso é que mulheres fumantes tem maior correlação do que hormens fumantes. A taxa de mortalidade por doenças cardio vasculares ou proporção de pessoas com diabetes não teve correlação significativa com o total de mortes por milhão.

```
# Fazendo mapa de correlação

plt.subplots(figsize=(15,15))
sns.heatmap(corr, cmap='RdBu', fmt='0.01f', square=True, linecolor='white', annot=True);
```

```
# importando o dataset COVID-19 para um Data Frame
DATA_PATH = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
df = pd.read_csv(DATA_PATH,
                 parse_dates=True, index_col='date')
df.index
```

```
# Identificar os tipos de variáveis
df_brazil.dtypes
```

```
# Localizar a data em que acontece a primeira morte no Brasil.

df_brazil.loc[df_brazil.total_deaths == 1]
```

```
# Calcular quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.

from datetime import date
f_date = date(2019, 12, 31)
l_date = date(2020, 3, 18)
delta = l_date - f_date
print(delta.days)
```

```
fig, ax = plt.subplots(figsize=(15,10))

df_brazil.loc['2020-03-18':'2020-06-21',"new_deaths"].plot(kind="line", ax=ax)
ax.set_title("Número de mortes")
ax.set_ylabel("Número de novas mortes")
ax.set_xlabel("Data")

plt.tight_layout()
```

```
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,10))
df_brazil.loc['2020-02-26':'2020-06-21',"total_cases"].plot(kind="line", ax=ax1)
ax1.set_title("Gráfico de casos de COVID-19")
ax1.set_ylabel("Casos Confirmados")
ax1.set_xlabel("Data")

df_brazil.loc['2020-02-26':'2020-06-21':,"total_deaths"].plot(kind="line", ax=ax2)
ax2.set_title("Gráfico de mortes")
ax2.set_ylabel("Mortes Confirmados")
ax2.set_xlabel("Data")

plt.tight_layout()
```
